## Union datos sociodemográficos Madrid con geo
La idea es obtener un CSV con el CSV formateado y con una columna que sea el id de geo de la zona correspondiente.

```python
data_sociodem = pd.read_csv("./barrios_sociodem.csv")
data_sociodem.head()
```

```python
data_geo = pd.read_csv("./geo_elements.csv")
data_geo.head()
```

```python
# Comprobamos cuantos elementos tiene cada uno
data_geo = data_geo[data_geo.level_id == 3] 
len(data_geo.name.unique())
```

```python
len(data_sociodem.geo_barrio.unique())
```

```python
def get_geo_id(name):
    try:
        return data_geo[data_geo.name == name].id.values[0]
    except:
        print(name)

# Intentamos machear directamente por nombre, sin difflib
data_sociodem["geo_id"] = data_sociodem["geo_barrio"].apply(lambda x: get_geo_id(x))
data_sociodem.head()
```

```python
# Vamos a revisar el que no tiene valor
data_sociodem[data_sociodem["geo_barrio"].isna()]
```

```python
# El retiro no tiene valor por algun motivo. Vamos a ver cual se ha quedado huerfano
data_geo[~data_geo["name"].isin(data_sociodem.geo_barrio.unique())]
```

```python
# Macheamos a mano
data_sociodem.loc[data_sociodem["dem_CodBarrio"] == 31, "geo_barrio"] = "Pacífico"
data_sociodem.loc[data_sociodem["dem_CodBarrio"] == 31, "geo_id"] = 70
data_sociodem.head()
```

```python
# Formateamos DF para mayor facilidad a la hora de cargar en BBDD
data_sociodem = data_sociodem.drop(['dem_CodBarrio', 'dem_Barrio', 'geo_barrio'], axis=1)
data_sociodem.head()
```

```python
data_sociodem.columns = ["pob_total", "count_households", "avg_size_home", "unemployment_rate", "prop_without_studies", "prop_without_university_studies", "prop_with_university_studies", "density", "avg_age", "index_youth", "index_dependency", "index_active_population", "index_replacement_active_population", "prop_born_outside", "geo_id"]
data_sociodem.head()
```

```python
data_sociodem.to_csv("./SocioDem_Geo.csv")
```
