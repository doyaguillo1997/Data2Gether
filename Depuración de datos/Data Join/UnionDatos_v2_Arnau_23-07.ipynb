## Datos Idealista

```python
data_idealista = pd.read_csv("./Madrid_Idealista_Variable_Seleccionadas.csv")
barrios_idealista = data_idealista.neighborhood.unique()
```

```python
idealista_geo_names = {}
names_not_found = []
for name in barrios_idealista:
    similar = difflib.get_close_matches(name, barrios_geo, cutoff=0.7)
    if similar:
        idealista_geo_names[name] = similar[0]
    else:
        names_not_found.append(name)
        
len(idealista_geo_names)
```

```python
idealista_geo_names
```

```python
# Vamos a ver como cuadra los nombres restantes con un punto de corte menor
idealista_geo_names_not_found = {}

for name in names_not_found:
    similar = difflib.get_close_matches(name, barrios_geo, cutoff=0.4)
    if similar:
        idealista_geo_names_not_found[name] = similar[0]
    else:
        print(name)
        
idealista_geo_names_not_found
```

```python
idealista_geo_names['Ventilla-Almenara'] = 'Almenara'
idealista_geo_names['Tres Olivos - Valverde'] = 'Valverde'
idealista_geo_names['Nuevos Ministerios-Ríos Rosas'] = 'Rios Rosas'
idealista_geo_names['Conde Orgaz-Piovera'] = 'Piovera'
idealista_geo_names['Valdebebas - Valdefuentes'] = 'Valdefuentes'
idealista_geo_names['Chueca-Justicia'] = 'Justicia'
idealista_geo_names['Huertas-Cortes'] = 'Cortes'
idealista_geo_names['12 de Octubre-Orcasur'] = 'Orcasur'
idealista_geo_names['El Cañaveral - Los Berrocales'] = 'El Cañaveral'
idealista_geo_names['Valdebernardo - Valderribas'] = 'Valdebernardo'
idealista_geo_names['Campo de las Naciones-Corralejos'] = 'Corralejos'
```

```python
not_found_names_idealista = ['Arroyo del Fresno',
'Las Tablas',
'Montecarmelo',
'Sanchinarro',
'Virgen del Cortijo - Manoteras',
'San Andrés',
'Ambroz',
'Pau de Carabanchel']
```

```python
data_idealista["geo_name"] = data_idealista["neighborhood"].apply(lambda x: idealista_geo_names[x] if x in idealista_geo_names else None)
data_idealista.head()
```

```python
# Duplicados  
print(data_idealista.duplicated().sum())
# Nulos
print(data_idealista.isnull().sum())
```

```python
data_idealista.drop(columns=['Unnamed: 0', 'propertyType', 'parkingIncludedInPrice', 'parkingPrice'], inplace = True)
data_idealista.to_csv("idealista_data.csv", index = False)
```

# Unión 

```python
ide = pd.read_csv("idealista_data.csv")
hist = pd.read_csv('historico_abril.csv')
dem = pd.read_csv('sociodem_data.csv')
```

```python
# Añadimos prefijos a los datasets.
ide = ide.add_prefix("ide_")
hist = hist.add_prefix("his_")
dem = dem.add_prefix("dem_")
```

```python
ide.rename(columns={"ide_geo_name": "geo_name"}, inplace = True)
hist.rename(columns={"his_geo_name": "geo_name"}, inplace = True)
dem.rename(columns={"dem_geo_name": "geo_name"}, inplace = True)
```

```python
ide_hist = ide.merge(hist, on='geo_name')
```

```python
ide_hist_dem = ide_hist.merge(dem, on = 'geo_name')
```

```python
# Duplicados  
print(ide_hist_dem.duplicated().sum())
# Nulos
print(ide_hist_dem.isnull().sum())
```

```python
# Eliminamos 146 duplicados
ide_hist_dem.drop_duplicates(subset=['ide_floor', 'ide_size', 'ide_address', 'ide_district', 'ide_neighborhood', 'ide_latitude', 'ide_longitude', 'ide_showAddress','ide_hasParking'], inplace = True)
```

```python
ide_hist_dem.to_csv('ide_hist_dem.csv', index = False)
```
