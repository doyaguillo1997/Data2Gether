```python
import pandas as pd
import os
```

## Tratamiento Distribución
La finalidad de esto es tener un único fichero con la información de todos los barrios. Para ello se usan todos los ficheros generados por el script de WebScrapping de Idealista.

```python
files = os.listdir()
# Get districti distributions. Distribution file has the neigborhood information about the district
distribution_files = [file for file in files if "distribución" in file and not "madrid" in file]
len(distribution_files)
```

```python
data = pd.read_csv(distribution_files[0], sep="\t")
data
```

```python
data_joined = pd.DataFrame()
# Read files
for file in distribution_files:
    data = pd.read_csv(file, sep="\t")
    data["Distrito"] = file.split('_')[0]
    # Join all data
    data_joined = data_joined.append(data.drop(["Unnamed: 0"], axis=1))
    

data_joined = data_joined.reset_index(drop=True) # Remove duplicated indexes
data_joined = data_joined.rename(columns={"Localización": "Barrio"})
data_joined
```

```python
# Save CSV
data_joined.to_csv("./Procesados/resumen_barrios.csv")
```

## Tratamiento historicos
En este apartado la idea es agrupar los historicos de Madrid municipio y sus distritos. Evitando asi tener un fichero para cada una de las zonas.

```python
historic_files = [file for file in files if "historico" in file]
len(historic_files)
```

```python
data = pd.read_csv(historic_files[0], sep="\t")
data
```

```python
data_joined = pd.DataFrame()
for file in historic_files:
    data = pd.read_csv(file, sep="\t")
    data["Zona"] = file.split('_')[0]
    # Eliminamos la columna Unnamed: 0 y guardamos
    data_joined = data_joined.append(data.drop(["Unnamed: 0"], axis=1))
    
data_joined = data_joined.reset_index(drop=True)
data_joined
```

```python
data_joined.to_csv("./Procesados/historico_madrid.csv")
```
